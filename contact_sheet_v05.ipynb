```
# Concrete Experiment

filePaths = [
             "/content/local_drive/concept1_in/all/_DSC8208.jpg",
             "/content/local_drive/concept1_in/all/_DSC8205.jpg",
             "/content/local_drive/concept1_in/all/_DSC8192.jpg",
             "/content/local_drive/concept1_in/all/_DSC8201.jpg",
             "/content/local_drive/concept1_in/all/_DSC8213.jpg",
            "/content/local_drive/concept1_in/all/_DSC8443.jpg",
             #"/content/local_drive/concept1_in/all/_DSC8457.jpg",
             ]
result = None

WIDTH = 5000
HEIGHT = 3000

for f in filePaths:
    assert os.path.isfile(f)
    print(f)
    img = cv2.imread(f).astype(np.float32)
    #print(img.shape)
    cropped = img[0:HEIGHT, 0:WIDTH]
    rows, cols, x = cropped.shape
    print("Cropped: ", cropped.shape )

    INCREMENTS = 50
    r2 = (INCREMENTS * len(filePaths)) * np.ones(shape=cropped.shape, dtype=np.float32)
    if result is None:
        result = np.zeros(shape=cropped.shape, dtype=np.float32)

    rotationScale = 0.025
    halfTotalRotation = (360 * rotationScale) / 2
    
    print("Rotations")
    for i in range(INCREMENTS):
        #print((360/INCREMENTS) * i)
        #print((cols-1)/2.0)
        rotation = (((360/INCREMENTS) * i) * rotationScale) - halfTotalRotation
        #print(rotation)
        m = cv2.getRotationMatrix2D(((WIDTH/INCREMENTS)*i, (rows-1)/2.0), rotation, 1)
        rotated = cv2.warpAffine(cropped, m, (cols, rows))
        r = result
        #print(r.shape)
        #print(rotated.shape)
        result = cv2.add(r, rotated)



#for i in images:
#    r = result
#    result = cv2.add(r, i)

finalResult = cv2.divide(result, r2)    
imgPath = outPath2 + "/{0}_f6.jpg".format("concrete_experiment_v100")
fImg = finalResult.astype(np.uint8)
print(imgPath)
cv2.imwrite(imgPath, fImg)
!cp $imgPath /content/drive/My\ Drive/Documents/ContactSheet/src/

```
