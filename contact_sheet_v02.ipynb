```
# This is the cell that generated the actual files
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os

path = r"/content/local_drive/f6Monteray/"
outPath = path + "fft_1"

print("inpath: {0}".format(path))
print("outpath: {0}".format(outPath))

if os.path.isdir(outPath):
    pass
else:
    os.makedirs(outPath)

ROOT_DIR = os.path.abspath(path)
files = os.listdir(path)

for _file in files[:3]:
    filePath = ROOT_DIR + "/" + _file

    if os.path.isfile(filePath):
        print(filePath)
        lin_filePath = outPath + "/" + "lin_" + _file
        fft_filePath = outPath + "/" + "fft_" + _file
        fft2_filePath = outPath + "/" + "fft2_" + _file
        fft3_filePath = outPath + "/" + "fft3_" + _file
        ifft_filePath = outPath + "/" + "ifft3_" + _file

        img = cv2.imread(filePath,0)
        linear = np.array(255*(img / 255) ** 2.4, dtype = 'float32')
        f = np.fft.fft2(linear)
        fshift = np.fft.fftshift(f)


        dft_dft = np.fft.fft2(f)
        spectrum = np.abs(f)
        #c = 255/np.log(1 + np.amax(spectrum))
        #spectrum = c * np.log(spectrum)

        double_dft_spectrum = np.abs(dft_dft)
        c = 255 / np.log(1 + np.amax(double_dft_spectrum))
        #double_dft_spectrum = c * np.log(double_dft_spectrum)


        fshift2 = np.fft.fftshift(dft_dft)

        #magnitude_spectrum = 20*np.log(np.abs(fshift))
        magnitude_spectrum = np.abs(fshift)
        #magnitude_spectrum2 = 20*np.log(np.abs(spectrum))
        magnitude_spectrum3 = 10*np.log(np.abs(fshift))

        f_ishift = np.fft.ifftshift(fshift2)

        img_back = np.fft.ifft2(f_ishift)
        img_back = np.abs(img_back)

        print(fft_filePath)
        print(fft2_filePath)
        print(fft3_filePath)
        print(ifft_filePath)

        cv2.imwrite(lin_filePath, linear)
        cv2.imwrite(fft_filePath, magnitude_spectrum)
        cv2.imwrite(fft2_filePath, double_dft_spectrum)
        cv2.imwrite(fft3_filePath, magnitude_spectrum3)
        cv2.imwrite(ifft_filePath, img_back)

        plt.subplot(131),plt.imshow(magnitude_spectrum, cmap = 'gray')
        plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
        plt.subplot(132),plt.imshow(img_back)
        plt.title('shift back'), plt.xticks([]), plt.yticks([])
        plt.show()

        !cp {lin_filePath}  /content/drive/My\ Drive/Documents/ContactSheet/src/f6Monteray/fft_1
        !cp {fft_filePath}  /content/drive/My\ Drive/Documents/ContactSheet/src/f6Monteray/fft_1
        !cp {fft2_filePath}  /content/drive/My\ Drive/Documents/ContactSheet/src/f6Monteray/fft_1
        !cp {fft3_filePath}  /content/drive/My\ Drive/Documents/ContactSheet/src/f6Monteray/fft_1
        #plt.savefig(fft_filePath)    
```
